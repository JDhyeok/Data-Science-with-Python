```python
import pandas as pd
dataset = '../Data/Banking_Marketing.csv.'
df = pd.read_csv(dataset, header=0)
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.dropna(axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
df['education'].value_counts()
```

```python
df.education.replace({"basic.9y":"Basic","basic.6y":"Basic","basic.4y":"Basic"},inplace=True)
```

```python
import numpy as np
data_column_category = df.select_dtypes(exclude=[np.number]).columns
data_column_category

data_column_numeric = df.select_dtypes(include=[np.number]).columns
data_column_numeric
```

```python
from sklearn.preprocessing import StandardScaler
std_scale = StandardScaler()

df[data_column_numeric] = std_scale.fit_transform(df[data_column_numeric])
```

```python
df_onehot_category_frame = pd.get_dummies(df[data_column_category])
final_encoded_df = pd.concat([df[data_column_numeric],df_onehot_category_frame],axis=1)
```

```python
final_encoded_df.head()
```

```python
final_encoded_df.info()
```

```python
#Segregating Independent and Target variable
X=final_encoded_df.drop(columns='y')
y=final_encoded_df['y']

from sklearn. model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
print("FULL Dateset X Shape: ", X.shape )
print("Train Dateset X Shape: ", X_train.shape )
print("Train Dateset y Shape: ", y_train.shape )
print("Test Dateset X Shape: ", X_test.shape )
print("Test Dateset y Shape: ", y_test.shape )
```
