```python
import numpy as np
x = np.linspace(0,10,20)
print(x)
y = x**3
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots()
plt.show()
```

```python
fig, axes = plt.subplots()

axes.plot(x,y)
axes.plot(x,y,'D-k')
axes.set_xlabel('Linearly Spacd Numbers')
axes.set_ylabel('y Value')
axes.set_title('As x increases, y increases by x cubed')
plt.show()
```
