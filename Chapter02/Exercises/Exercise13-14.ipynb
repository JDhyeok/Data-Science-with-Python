```python
import numpy as np

x = np.linspace(0,10,20)
```

```python
y = x**3
```

```python
import matplotlib.pyplot as plt

plt.plot(x,y)
plt.show()
```

```python
plt.xlabel('Linearly spaced Nubmers')
plt.ylabel('y Value')
plt.title('x by x Cubed')
plt.plot(x,y,'k')
```

```python
plt.plot(x,y,'Dk')
```

```python
plt.plot(x,y,'D-k')
```

```python
y2 = x**2
plt.plot(x,y,'D-k',label='x cubed')
plt.plot(x,y2,'--r',label='x squared')
plt.legend(loc='upper left')
```

```python
plt.figure(figsize=(10,5))
plt.plot(x,y,'D-k',label='x cubed')
plt.plot(x,y2,'--r',label='x squared')
plt.legend(loc='upper left')
```
