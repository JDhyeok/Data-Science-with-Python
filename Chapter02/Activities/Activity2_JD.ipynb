```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_list = ['Jan','Feb','Mar','Apr','May','Jun']
y_list = [1000,1200,1400,1600,1800,2000]
```

```python
plt.plot(x_list,y_list,'*:b')
plt.xlabel('Month')
plt.ylabel('Items Sold')
plt.title('Items Sold has been Increasing Linearly',fontsize=22)
plt.show()
```
